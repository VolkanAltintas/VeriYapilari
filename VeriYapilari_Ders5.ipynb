```python
def histogram(deger):
    hist_dict = {}
    for k in deger:  # k = karakter k=i
        if k in hist_dict: # hist_dict={'B':1, 'i':2, 'l':1,'g':1}
            hist_dict[k] += 1
        else:
            hist_dict[k] = 1
    return hist_dict
 
print (histogram("Bilgisayar"))
```

```python
def numara_bul(dict,numara):
    for i in dict.values():  # Srasyla dictionary deerlerini dndryoruz.
        if i==numara or i==str(numara):  # gelen deeri numaramz ile kyaslyoruz.
            return True
    return False
```

```python
numbers = range(11)
new_dict_for = {}

for n in numbers:
    if n%2==0:
        new_dict_for[n] = n**2

print(new_dict_for)
```

```python
new_dict_comp = {n:n**2 for n in numbers if n%2 == 0}

print(new_dict_comp)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

dict1_cond = {k:v for (k,v) in dict1.items() if v>2}

print(dict1_cond)
```

```python
dict1_doubleCond = {k:v for (k,v) in dict1.items() if v>2 if v%2 == 0}
print(dict1_doubleCond)
```

```python
dict1 = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f':6}

dict1_tripleCond = {k:('ift' if v%2==0 else 'tek') for (k,v) in dict1.items()}

print(dict1_tripleCond)
```

**Dictionaries for text analysis**

```python
def wordcount(fname):
       try:
           fhand=open(fname)
       except:
           print('Dosya alamad...')
           exit()
       count= dict()
       for line in fhand: #line =satr
           words = line.split() #cmleleri kelimelere blme
           for word in words:
               if word not in count:
                   count[word] = 1
               else:
                   count[word] += 1
       return(count)
```
